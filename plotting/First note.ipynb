# Python"Functions":
Functions are a reuseble, organized code to do a specific tasks. in Python packages, there are many built-in functions like "print()","exp()", "Sqrt()", etc. 
You also can write your own fuction which is called, user-defined fuctions. 

This is how you define a function in Python: 

1. Function blocks begin with the keyword def followed by the function name and parentheses ( ( ) ).
2. Any input parameters or arguments should be placed within these parentheses. You can also define parameters inside these parentheses.
3. The code block within every function starts with a colon (:) and is indented.
4. The statement return [expression] exits a function, optionally passing back an expression to the caller. A return statement with no arguments is the same as return None.

Let's practice that:

```python
def first_function():
    x = 3 
    y = x**3 
    z = np.sqrt(y)
    return(y, z)

first_result = first_function()
print(first_result)
```

# Making a scatter plot: 
To visuallize the data using the scatter plot. We will use some random points.

matplotlib.pyplot.scatter(x_axis_data, y_axis_data, s=None, c=None, marker=None, cmap=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None) 

The scatter() method takes in the following parameters: 
 

- x_axis_data- An array containing x-axis data
- y_axis_data- An array containing y-axis data
- s- marker size (can be scalar or array of size equal to size of x or y)
- c- color of sequence of colors for markers
- marker- marker style
- cmap- cmap name
- linewidths- width of marker border
- edgecolor- marker border color
- alpha- blending value, between 0 (transparent) and 1 (opaque)
- Except x_axis_data and y_axis_data all other parameters are optional and their default value is None. Below are the scatter plot examples with various parameters.

```python
#import matplotlib.pyplot as plt # We already importet this on top

fig, ax1 = plt.subplots(1,1,figsize=(12,6), dpi= 80)
# dataset-1
x1 = [22, 24, 56, 92, 105]
 
y1 = [15, 100, 95, 89, 200]
 
# dataset2
x2 = [4, 15, 200, 98, 50, 16]
 
y2 = [40, 35, 80, 17, 25, 30]
 
plt.scatter(x1, y1, c ="yellow", linewidths = 2, marker ="s",edgecolor ="green",s = 50)
         
plt.scatter(x2, y2, c ="red",linewidths = 2,marker ="^", edgecolor ="blue",  s = 100)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend(["x1: {}".format(x1),"x2: {}".format(x2)])

plt.show()
```

# Using For Loops: 
There is a sample code for making 4subplots in one plot, using the for loop. Here, we also used the random numbers.There are some more examples to use at the end of these notes to how to space your plots using grid!

```python
# Draw multiple plots using for-loops:
#import numpy as np
from numpy.random import seed, randint #Using random numbers
seed(100) 

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=True, dpi=120)

# Define the colors and markers to use
colors = {0:'b', 1:'r', 2:'y', 3:'g'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Plot each axes using for loop: 
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,20,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('plot: ' + str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('How to plot 4subplots in one figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
plt.show()
```

# Some useful plots: 
Here are the examples of useful plots that you can use it for your assignment. These are just some examples so if you need more info on how to plot each of these, you can find more information here: [Tutorials](https://matplotlib.org/stable/tutorials/index.html)

```python
# Setup the subplot2grid Layout
fig = plt.figure(figsize=(10, 5))
ax1 = plt.subplot2grid((2,4), (0,0)) 
ax2 = plt.subplot2grid((2,4), (0,1)) 
ax3 = plt.subplot2grid((2,4), (0,2)) 
ax4 = plt.subplot2grid((2,4), (0,3)) 
ax5 = plt.subplot2grid((2,4), (1,0), colspan=2) 
ax6 = plt.subplot2grid((2,4), (1,2)) 
ax7 = plt.subplot2grid((2,4), (1,3)) 

# Input Arrays
n = np.array([0,2,4,6,8,10])
x = np.linspace(0,5,10)
xx = np.linspace(-0.5, 1., 100)

# Scatterplot
ax1.scatter(xx, xx + np.random.randn(len(xx)))# adding some random errors to the data
ax1.set_title("Scatter Plot")

# Step Chart
ax2.step(n, n**2, lw=2)
ax2.set_title("Step Plot")

# Histogram
ax3.hist(xx + np.random.randn(len(xx)))
ax3.set_title("Histogram")

# Fill Between
ax4.fill_between(x, x**2, x**4, color="steelblue", alpha=0.5);
ax4.set_title("Fill Between");

# Time Series
dates = pd.date_range('2022-01-01', periods = len(xx))
ax5.plot(dates, xx + np.random.randn(len(xx)))
ax5.set_xticks(dates[::30])
ax5.set_xticklabels(dates.strftime('%Y-%m-%d')[::30])
ax5.set_title("Time Series")

# Box Plot
ax6.boxplot(np.random.randn(len(xx)))
ax6.set_title("Box Plot")

# Bar Chart
ax7.bar(n, n**2, align="center", width=0.5, alpha=0.5)
ax7.set_title("Bar Chart")

fig.tight_layout()
```
